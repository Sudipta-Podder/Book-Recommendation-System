```python
#Pivot Table
book_pivot = final.pivot_table(index='user_id', columns='title',values='rating')
```

```python
from scipy.sparse import csr_matrix
```

```python
book_pivot.index
```

```python
book_matrix=csr_matrix(book_pivot.values)
```

```python
#Creating matrix
from sklearn.neighbors import NearestNeighbors
```

```python
#KNN Model
model_knn = NearestNeighbors(metric = 'cosine')
model_knn.fit(book_matrix)
```

```python
popularity_threshold = 100
popular_books = final[final['num_of_rating'] >= popularity_threshold]

print("Total Popular Books: ", len(popular_books['title'].unique()))
print(popular_books.head())
```

```python
def get_recommends(user_id="", k=10):
  recommended_books = [user_id, []]

  if user_id not in book_pivot.index:
        print(f"User ID {user_id} not found in the dataset.")
        return recommended_books

  user_row = book_pivot.loc[user_id].values.reshape(1, -1)
  distances, user_info = model_knn.kneighbors(user_row, k + 1, return_distance=True)

  recom_user_info = book_pivot.index[np.flip(user_info[0])[:-1]].to_list()

  recommended_books[1] = recom_user_info

  return recommended_books
```

```python
def predict_book_rating(user_id, isbn, model_knn, book_pivot, book_matrix, k=5, n= 11):
    # Find the book title corresponding to the given ISBN
    book_title = final[final['ISBN'] == isbn]['title'].iloc[0]

    # Check if the specified user ID is in the columns of the user-item matrix
    if user_id in book_pivot.index:

        # Print the book title
        print(f"Book title for ISBN {isbn}: {book_title}")
        # Find the index of the specified user ID in the user-item matrix
        query_index_user = book_pivot.index.get_loc(user_id)

        # Find the index of the specified book title in the user-item matrix
        query_index_book = book_pivot.columns.get_loc(book_title)

        # Use the trained model to find k-nearest neighbors
        distances, indices = model_knn.kneighbors(book_matrix[query_index_book, :].reshape(1, -1), n=k)

        # Get the ratings of the k-nearest neighbors for the specified book
        neighbor_ratings = book_matrix[indices.flatten(), query_index_user]

        # Predict the rating by taking the average of the neighbor ratings
        predicted_rating = np.mean(neighbor_ratings)

        # Print the predicted rating
        print(f"Predicted rating for user {user_id} and book {isbn}: {predicted_rating}")
        return predicted_rating
    else:
        print(f"User ID {user_id} not found in the dataset.")
        return None
```

```python
user_id_input = input("Enter a user ID: ")
isbn_input = input("Enter an ISBN number: ")

try:
    # Convert user ID input to integer
    user_id_input = int(user_id_input)

    # Call the predict_book_rating function
    predicted_rating = predict_book_rating(user_id_input, isbn_input, model_knn, book_pivot, book_matrix)

    if predicted_rating is None:
        print("Prediction not available.")

except ValueError:
    print("Invalid input. Please enter a valid user ID.")
    #277427
    #002542730X
```
